### PDF LOADER

```python
from langchain_community.document_loaders import (PyPDFLoader, PyMuPDFLoader)
```

```python
try:
    py_pdf_loader = PyPDFLoader("data/pdf/AIPdf.pdf")
    documents = py_pdf_loader.load()
    print(len(documents))
    for i, doc in enumerate(documents):
        print(f"Document {i+1}:\n{doc.page_content}\n")
except Exception as e:
    print(f"Error loading PDF with PyPDFLoader: {e}")
```

```python
py_mu_pdf_loader = PyMuPDFLoader("data/pdf/AIPdf.pdf")
documents = py_mu_pdf_loader.load()
print(len(documents))
for i, doc in enumerate(documents):
    print(f"Document {i+1}:\n{doc.page_content}\n")
```
