### Embedding Using OPENAI

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY") 
```

```python
from langchain_openai import OpenAIEmbeddings
embeddingOpenAI = OpenAIEmbeddings(model="text-embedding-3-small")
embeddingOpenAI
```

```python
sample_text = "This is a sample text to generate embeddings."
embedding_vector = embeddingOpenAI.embed_query(sample_text)
print(len(embedding_vector))
```

```python
multi_texts = ["This is the first sample text.", "This is the second sample text.", "This is the third sample text."]
embedding_vectors = embeddingOpenAI.embed_documents(multi_texts)
print(len(embedding_vectors[0]))
print(embedding_vectors)
```
