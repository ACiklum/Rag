### Create  Text File

```python
import os
os.makedirs("data/text_file", exist_ok=True)

sampleText = {
    "data/text_file/sample1.txt": """
    This is the first sample text file. It contains some example content for testing purposes.""",
    "data/text_file/sample2.txt": """
    This is the second sample text file. It also contains example content for testing purposes.""",
}

for file_path, content in sampleText.items():    
    with open(file_path, "w", encoding="utf-8") as f:
        f.write(content.strip())
```

### Read Text File

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader("data/text_file/sample1.txt", encoding="utf-8")
documents = loader.load()
print(documents[0].page_content)
```

### Read Multiple Files

```python
from langchain_community.document_loaders import DirectoryLoader

loader = DirectoryLoader("data/text_file",
                         glob="*.txt",
                         loader_cls=TextLoader,
                         show_progress=True,
                         )
documents =loader.load()
for doc in documents:
    print(doc.page_content)
```

### Text Splitter   

```python
from langchain_text_splitters import ( RecursiveCharacterTextSplitter, CharacterTextSplitter, TokenTextSplitter)
```

```python
char_splitter = CharacterTextSplitter(separator="\n",chunk_size=200, chunk_overlap=10, length_function=len)
texts = char_splitter.split_text(documents[2].page_content)
# print(documents[2].page_content)
print("\n--- Split Texts ---\n")
print(len(texts))
for i, text in enumerate(texts):
    print(f"Chunk {i+1}:\n{text}\n")



```

```python
rescursive_splitter = RecursiveCharacterTextSplitter(
    separators=["\n\n", "\n", " ", ""], chunk_size=400, chunk_overlap=50, length_function=len)

rescursive_split_text = rescursive_splitter.split_text(documents[2].page_content)

for i, data in enumerate(rescursive_split_text):
    print(f"Chunk {i+1}:\n{data}\n")
```

```python
token_splitter = TokenTextSplitter(chunk_size=200, chunk_overlap=20)
token_split_text = token_splitter.split_text(documents[2].page_content)
for i, data in enumerate(token_split_text):
    print(f"Chunk {i+1}:\n{data}\n")
```
